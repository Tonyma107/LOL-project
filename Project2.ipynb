# Part 4 Exploratory Data Visualization Data Science:

#### Can we predict the total damage taken from the damage mitigated per minutes?

I want to see whether if you mitigated more damage, and will you be taken more damage. So I would Generate a scatter plot from `df1` where the dependent variable ($y$) is `damage per minutes` and independent variable ($X$) is `damagemitigatedperminute`. And I will be utilizing a linear regression model to answer this question.  Damage per minutes means the damages a champion is doing to the enemy team's champion only. 


```python
plt.scatter(df1['damagemitigatedperminute'], df1['dpm'])
```

### What is the residual or the correlation?

```python
r = df1['dpm'].corr(df1['damagemitigatedperminute'])
r
```

This correlation is very close to one and we can see from the scatter plot that the relationship is  positive and the linear regression line seems to be very close to be linear and the relationship is  strong. This suggests that as damages per minute increases, so does the mitigated damage.

```python
sdy = df1['dpm'].std()
sdx= df1['damagemitigatedperminute'].std()
slope = r * (sdy / sdx)
slope
```

This means that For every one-unit increase in damage mitigated per minute, the predicted change in damage  per minute is equal to the slope.

```python
avgy = df1['dpm'].mean()
avgx = df1['damagemitigatedperminute'].mean()
intercept = avgy - (slope * avgx)
intercept
```
